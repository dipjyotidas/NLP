### Live Twitter Sentiment Analysis

* This script can be used to do Live Twitter sentiment analysis on any topic.
* example - stock market, car etc. from a tweet !

#### Import all the libaries and the API tokens from the Twitter developer website

From tweets, we can easily pass it through our sentiment_analysis module!

```python
# Import all the libaries for connecting to Twitter's API

from tweepy import Stream
from tweepy import OAuthHandler
from tweepy.streaming import StreamListener
import json
import sentiment_analysis as s


# Get the consumer key, consumer secret, access token, access secret from a Twitter Developer account :

ckey = "XXXXXXXXXXXXXXXXXXXXXXX"
csecret = "XXXXXXXXXXXXXXXXXXXXXXX"
atoken = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
asecret = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"


# This function will return the sentiment of a particular word, topic, example- 'stock market', 'car','happy new year' etc from tweets!

class listener(StreamListener):

    def on_data(self, data):

        all_data = json.loads(data)

        tweet = all_data["text"]
        
        sentiment_value, confidence = s.sentiment(tweet) # we want to gather the sentiment on the tweet (pos or neg and the confidence level)
        # using 5 algorithms - to get the confidence level
        print(tweet, sentiment_value, confidence)

        if confidence*100 >= 80: # if there are 4 or more votes in favor of sentiment value - pos or neg
            output = open("twitter-out.txt","a") # saving it on a .txt file
            output.write(sentiment_value)
            output.write('\n') # new line
            output.close()

        return True

    def on_error(self, status):
        print(status)

auth = OAuthHandler(ckey, csecret)
auth.set_access_token(atoken, asecret)

twitterStream = Stream(auth, listener())
twitterStream.filter(track=["stock market"])

# Some of the tweets with the topic - 'stock market' are printed below :
```

##### Plotting the sentiment of the topic - 'stock market' :

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.animation as animation
from matplotlib import style
import time

style.use("ggplot")

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    pullData = open("twitter-out.txt","r").read()
    lines = pullData.split('\n') # new line element

    xar = [] # x array
    yar = [] # y array

    x = 0
    y = 0

    for l in lines[-200:]: # for the last 200 lines
        x += 1
        if "pos" in l: # if line is pos
            y += 1
        elif "neg" in l:
            y -= 1 # we can modify the neg sentiment using different numbers, eg -0.3

        xar.append(x)
        yar.append(y)
        
    ax1.clear()
    ax1.plot(xar,yar)
ani = animation.FuncAnimation(fig, animate, interval=1000)
plt.show()
```

* This implies negative sentiment of the word - 'stock market' from the tweets.
